```python
# Q1: What categories of passengers were most likely to survive the Titanic disaster? (2 points) 
#     The detailed explanation of the logic of the analysis (2 points)
#
# A1: It is a struggle in selection of attributes of where to get started. Inspired by the movie Titanic, at the sinking
#     scene where Woman and Child were allowed onboarding to life safing boat in higher priroty. Therefore, I selected "Sex" 
#     as my primary attribute in filtering, where "Parch" as the secondary and the remainings as n-ary.
#
#     The below calculation has demonstrated Female has a higher survival rate than Male. Therefore, in the remaining 
#     Q&A, I would focus on Female only.
#
#     There was a misconceptual calculation during the analysis and I corrected that afterwards.
#     Initially I adopted to calculate probability on each attribute, with value between 0 to 1. 
#     Then by picking some of the attributes under assumption to apply mutiplication to get the survival rate.
#     However, by doing so, there is no relation between attributes. For example 
#          survival rate of: women is 0.2, Parch(0) is 0.4
#                           = 0.2 * 0.04  
#                           = 0.08 (or 8%)
#     vs.
#          print("female + parch(0):", sum(df[ (df['Sex']=='female') & (df['Parch']==0) ]['Survived'])/totalNumWomenSurvived)
#
#     This is because the & condition links up the relation that restricted the search criteria (like SQL), where the 
#     probability way is more for random unrelated items to calcualte occurance rate.
#
#     My answer in Q2, will provide more details to answer Q1.

women = df.loc[df.Sex == 'female']["Survived"]
men = df.loc[df.Sex == 'male']["Survived"]
totalNumWomenSurvived = sum(women)
totalNumMenSurvived = sum(men)

print ("totol #survived women:", totalNumWomenSurvived)  #sum(women))
print ("totol #survived men:", totalNumMenSurvived)  #sum(men))
print ("total #passengers", myMaxNumPassenger)
rate_women = sum(women)/myMaxNumPassenger  #len(women)
rate_men = sum(men)/myMaxNumPassenger     #len(men)
print("rate of survived (women):", rate_women)
print("rate of survived (men):", rate_men)

```
